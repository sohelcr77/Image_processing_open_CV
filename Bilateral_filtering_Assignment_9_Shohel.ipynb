# Apply bilateral Filter on an image and explain differents parameters and result

```python
# Importing required library for processing

import cv2
import numpy as np
from matplotlib import pyplot as plt

#  storing image into image_path variable and Loading the image

 # I tried with different images to obtain better result
#image_path= "/content/drive/MyDrive/Computer Vision /111.jpeg"
#image_path= "/content/drive/MyDrive/Computer Vision /iphone.jpg"
image_path= "/content/drive/MyDrive/Computer Vision /images (46).jpeg"

image = cv2.imread(image_path)


# Convert the image to RGB for visualization
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Bilateral Filter with different parameters

''' Here Im trying different parameters of sigma color and sigma space values
    to find an optimal image. in first image i used low sigma color value to smooth pixels but preserve
    details ,on the other hand I used low sigma space to minimize the area of smoothing and preserving
    details and edges '''

filtered_1 = cv2.bilateralFilter(image, d=9, sigmaColor=50, sigmaSpace=50)

''' Here I tried to combine  Higher sigma  color value to  lower Sigma space and
    Lower Sigma color to Higher Sigma Space Thats resulting blending more colors to smooth pixels
    in wider area  preserving less  details and  slightly blurr edges '''

filtered_2 = cv2.bilateralFilter(image, d=9, sigmaColor=150, sigmaSpace=50)
filtered_3 = cv2.bilateralFilter(image, d=9, sigmaColor=50, sigmaSpace=150)

''' Here I tried to combine  Higher sigma  color value to  higherr Sigma space in multiple occations
    thats resulting image appears significantly smoothed out, with less texture and fewer details.
    There is a considerable reduction in noise This combination gives the smoothest a look,
    with a lot of detail loss and The edges are blurred, '''


filtered_4 = cv2.bilateralFilter(image, d=9, sigmaColor=150, sigmaSpace=150)
filtered_5 = cv2.bilateralFilter(image, d=9, sigmaColor=250, sigmaSpace=250)


# Converting  to RGB for displaying
filtered_1_rgb = cv2.cvtColor(filtered_1, cv2.COLOR_BGR2RGB)
filtered_2_rgb = cv2.cvtColor(filtered_2, cv2.COLOR_BGR2RGB)
filtered_3_rgb = cv2.cvtColor(filtered_3, cv2.COLOR_BGR2RGB)
filtered_4_rgb = cv2.cvtColor(filtered_4, cv2.COLOR_BGR2RGB)
filtered_5_rgb = cv2.cvtColor(filtered_5, cv2.COLOR_BGR2RGB)


# Displaying the results
plt.figure(figsize=(10, 12))

# Original image
plt.subplot(3, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Filtered images with different parameter combinations
plt.subplot(3, 2, 2)
plt.imshow(filtered_1_rgb)
plt.title('sigmaColor=50, sigmaSpace=50')
plt.axis('off')

plt.subplot(3, 2, 3)
plt.imshow(filtered_2_rgb)
plt.title('sigmaColor=150, sigmaSpace=50')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(filtered_3_rgb)
plt.title('sigmaColor=50, sigmaSpace=150')
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(filtered_4_rgb)
plt.title('sigmaColor=150, sigmaSpace=150')
plt.axis('off')

plt.subplot(3, 2, 6)
plt.imshow(filtered_5_rgb)
plt.title('sigmaColor=250, sigmaSpace=250')
plt.axis('off')


plt.tight_layout()
plt.show()
```
