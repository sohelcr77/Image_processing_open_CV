```python
Task-2:: Take three images of your choice and apply different image thresholding on them.

# I am using three imagese and applying different thresholding technique to them



import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image1 = cv2.imread("/content/drive/MyDrive/Computer Vision /images (1).jpeg", cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread("/content/drive/MyDrive/Computer Vision /images (2).jpeg", cv2.IMREAD_GRAYSCALE)
image3 = cv2.imread("/content/drive/MyDrive/Computer Vision /images (46).jpeg", cv2.IMREAD_GRAYSCALE)

# Applying different types of thresholding on image1,2,3

#1. Binary Thresholding on Image 1
ret,binary_thresh = cv2.threshold(image1, 160, 255, cv2.THRESH_BINARY)

# 2. Binary Inverse Thresholding on Image 1
ret,binary_inv_thresh = cv2.threshold(image1, 150, 225, cv2.THRESH_BINARY_INV)

# 3. Truncated Thresholding on Image 2
ret,trunc_thresh = cv2.threshold(image2, 127, 255, cv2.THRESH_TRUNC)

# 4. Threshold to Zero on Image 2
ret,tozero_thresh = cv2.threshold(image2, 50, 255, cv2.THRESH_TOZERO)

# 5. Threshold to Zero Inverse on Image 3
ret,tozero_inv_thresh = cv2.threshold(image3, 140, 255, cv2.THRESH_TOZERO_INV)

# 6 Binary Thresholding on Image 3
ret,binary_thresh_on_im3 = cv2.threshold(image3, 160, 255, cv2.THRESH_BINARY)

# Displaying images in a 3x2 grid
plt.figure(figsize=(15, 10))

# Binary Thresholding
plt.subplot(3, 2, 1)
plt.imshow(binary_thresh, cmap="gray")
plt.title("Binary Thresholding")

# Binary Inverse
plt.subplot(3, 2, 2)
plt.imshow(binary_inv_thresh, cmap="gray")
plt.title(" Binary Inverse")

# Truncated Thresholding
plt.subplot(3, 2, 3)
plt.imshow(trunc_thresh, cmap="gray")
plt.title(" Truncated Thresholding")

# Threshold to Zero
plt.subplot(3, 2, 4)
plt.imshow(tozero_thresh, cmap="gray")
plt.title(" Threshold to Zero")

# Threshold to Zero Inverse
plt.subplot(3, 2, 5)
plt.imshow(tozero_inv_thresh, cmap="gray")
plt.title("Threshold to Zero Inverse")

#  Binary Thresholding on Image 3
plt.subplot(3, 2, 6)
plt.imshow(binary_thresh_on_im3,cmap="gray")
plt.title("Binary on Image 3")




plt.tight_layout()
plt.show()
```
