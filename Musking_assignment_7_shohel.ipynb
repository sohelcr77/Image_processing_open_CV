Task-1 :: Take an image find LB and UB and make the binary mask for that colored object
Opt-Task-1 :: Apply the binary mask for task-1

```python
# Importing required libraries for image masking task-1


import cv2
import numpy as np
import matplotlib.pyplot as plt

# storing image into image_path variable

image_path= "/content/drive/MyDrive/Computer Vision /111.jpeg"
img = cv2.imread(image_path)

# since open cv reads image in BGR format,now converting BGR to RGB image format

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Now converting RGB image to HSV image for creating binary mask

hsv_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2HSV)

lower_blue =np.array([100, 170, 0])
upper_blue = np.array([145, 255, 255])

blue_mask = cv2.inRange(hsv_img, lower_blue, upper_blue)


plt.figure(figsize=(6,12))

plt.subplot(1,2,1)
plt.imshow(rgb_img)
plt.title(" RGB Image ")

plt.subplot(1,2,2)
plt.imshow(blue_mask, cmap='gray')
plt.title(" Blue Mask ")

plt.tight_layout()
plt.show()



```

#Take an image, define roi and make a binary mask for that image. :: Opt-Task-1 :: Apply the binary mask for task-1


```python
# loading a new image for making ROI and Masking on it

#defining image path
img_path2=  "/content/drive/MyDrive/Computer Vision /114.jpg"

#Loading the image
img_roi = cv2.imread(img_path2)

# making a black mask
mask_roi = np.zeros(img_roi.shape,np.uint8)

# Selecting the ROI
mask_roi[200:1000,140:750]=255

# applying ROI masking on orginal Image
final_masked_img = cv2.bitwise_and(img_roi,mask_roi)


''' Showing resulting Segmented Image side by side with Original Image
        so that we can make appropriate adjustment on ROI  '''


plt.figure(figsize=(6,12))

plt.subplot(1,2,1)
plt.imshow(img_roi)
plt.title(" ROI on Image ")

plt.subplot(1,2,2)
plt.imshow(final_masked_img)
plt.title(" Resulting Segmented Image ")

plt.tight_layout()
plt.show()
```
