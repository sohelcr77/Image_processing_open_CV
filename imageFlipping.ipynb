```python
# Task no 1--  Plot a 4x4(I found 2*2 grid is  more appropriare for the task ) grid with original and three flipped images.


import cv2
import matplotlib.pyplot as plt

# Load the image (I used drive mount image for processing)

image = cv2.imread('/content/drive/MyDrive/Computer Vision /images (1).jpeg')

# Flipping  the image in different according to flip code(>0, 0 , and <0)

flipped_vertically = cv2.flip(image, 0)
flipped_horizontally = cv2.flip(image, 5)
flipped_both = cv2.flip(image, -5)

# Displaying images in a 2x2 grid
plt.figure(figsize=(10, 10))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(image)
plt.title("Original Image")


# Vertically Flipped Image
plt.subplot(2, 2, 2)
plt.imshow(flipped_vertically)
plt.title("Flipped Vertically")

# Horizontally Flipped Image
plt.subplot(2, 2, 3)
plt.imshow(flipped_horizontally)
plt.title("Flipped Horizontally")

# Flipped Both Vertically and Horizontally
plt.subplot(2, 2, 4)
plt.imshow(flipped_both)
plt.title("Flipped Both Ways")


# Show the grid
plt.show()
```

```python
# Task-2:: Plot 2x3 grid with rotated images, each row containing images rotated by one of the two methods only

# Load the image
image = cv2.imread("/content/drive/MyDrive/Computer Vision /images (2).jpeg")



# Method-1
rotated_90_clockwise = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
rotated_90_counterclockwise = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
rotated_180 = cv2.rotate(image, cv2.ROTATE_180)



# Method 2

image_rot = cv2.imread("/content/drive/MyDrive/Computer Vision /images (37).jpeg")

# Get the dimensions of the image
(h, w) = image_rot.shape[:2]

# Define the center of the image
center = (w // 2, h // 2)

# Specify the rotation angle in degrees
angle1 = 90
angle2 = 180
angle3 = 270

# Get the rotation matrix
rotation_matrix1 = cv2.getRotationMatrix2D(center, angle1, 1.0)
rotation_matrix2 = cv2.getRotationMatrix2D(center, angle2, 1.0)
rotation_matrix3 = cv2.getRotationMatrix2D(center, angle3, 1.0)

# Perform the rotation
rotated_image1 = cv2.warpAffine(image_rot, rotation_matrix1, (w, h))
rotated_image2 = cv2.warpAffine(image_rot, rotation_matrix2, (w, h))
rotated_image3 = cv2.warpAffine(image_rot, rotation_matrix3, (w, h))

# Displaying images in a 2x3 grid
plt.figure(figsize=(12, 12))

# Method-1 90_clockwise
plt.subplot(2, 3, 1)
plt.imshow(rotated_90_clockwise)
plt.title("M1 90 Clockwise")

# Method-1 90_Counter_clockwise
plt.subplot(2, 3, 2)
plt.imshow(rotated_90_counterclockwise)
plt.title(" M1 90 Counter clockwise")

# M1 180 degree
plt.subplot(2, 3, 3)
plt.imshow(rotated_180)
plt.title(" M1 180 degree")

# 90 degree rotated image
plt.subplot(2, 3, 4)
plt.imshow(rotated_image1)
plt.title(" rotated 90 degree")

# 180 degree rotated image
plt.subplot(2, 3, 5)
plt.imshow(rotated_image2)
plt.title("rotated 180 degree")

#  270 degree rotated image
plt.subplot(2, 3, 6)
plt.imshow(rotated_image3)
plt.title("rotated 270 degree")




# Show the grid
plt.show()


```

```python
# Task-3:: plot original, re-scaled and resized images in a 1x3 grid


# Load the image
org_image = cv2.imread("/content/drive/MyDrive/Computer Vision /images (46).jpeg")

# Define the scaling factors
scale_x = 0.7  # Horizontal scale
scale_y = 0.7  # Vertical scale

# Rescale the image
rescaled_image = cv2.resize(org_image, None, fx=scale_x, fy=scale_y, interpolation=cv2.INTER_LINEAR)



# Define the target dimensions
width = 500
height = 500

# Resize the image
resized_image = cv2.resize(org_image, (width, height), interpolation=cv2.INTER_LINEAR)

#Displaying images in a 1x3 grid
plt.figure(figsize=(8, 24))

# Original Image

plt.subplot(1, 3, 1)
plt.imshow(org_image)
plt.title("Original Image")

# rescaled_image
plt.subplot(1, 3, 2)
plt.imshow(rescaled_image)
plt.title(" Rescaled Image")

# resized_image
plt.subplot(1, 3, 3)
plt.imshow(resized_image)
plt.title(" Resized Image")

# Show the grid
plt.show()

```
